### 2. 转换单个文档

````python
from docling.document_converter import DocumentConverter

converter = DocumentConverter()

# 支持本地路径或 URL
source = "https://arxiv.org/pdf/2408.09869"
# source = ""
result = converter.convert(source)
doc = result.document

# 导出为 Markdown
md_output = doc.export_to_markdown()
print(md_output[:100])
````

````python
import os

# 定义输出路径
output_dir = "D:\\"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    
file_name = os.path.basename(source).split('.')[0]
output_path = os.path.join(output_dir, f"{file_name}.md")

# 保存 Markdown 文件
with open(output_path, "w", encoding="utf-8") as f:
    f.write(md_output)
        
print(f"转换完成，文件已保存至: {output_path}")
````

````python
# 探索文档结构
for i, text_item in enumerate(doc.texts[:5]):
    print(f"[{i}] label={text_item.label}, text={text_item.text[:80]}")

for i, table in enumerate(doc.tables):
    print(f"表格{i}: {table.data.num_rows}x{table.data.num_cols}")

print(f"图片数量: {len(doc.pictures)}")
````
